# Agregación de datos por categoría

```python
import pandas as pd
import numpy as np
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle class", "Rich"]
```

```python
n = 500
gender_collection = []
category_collection = []

for i in range(0,n):
    gender_collection.append(np.random.choice(gender))
    category_collection.append(np.random.choice(income)) 
    
height_collection = 160 + 20*np.random.randn(n)
weight_collection = 65 + 25*np.random.randn(n)
age_collection = 30 + 12*np.random.randn(n)
income_collection = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame({
    "Gender" : gender_collection,
    "Category" : category_collection,
    "Height" : height_collection,
    "Weight" : weight_collection,
    "Age" : age_collection,
    "Income" : income_collection
})
data
```

## Agrupación de datos

```python
grouped_gender = data.groupby("Gender")
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)    
    print(groups)    
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender", "Category"])
double_group.groups
```

```python
double_group.get_group(("Female", "Poor"))
```

## Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group["Age"].describe()
```

```python
double_group.aggregate(
{
    "Income" : np.sum,
    "Age" : np.mean,
    "Height" : np.std
})
```

```python
double_group.aggregate({
    "Age" : np.mean,
    "Height" : lambda h: np.mean(h)/np.std(h)
})
```

```python
double_group.aggregate({np.sum, np.mean, np.std})
```

```python
double_group.aggregate({lambda x: np.mean(x) / np.std(x)})
```

## Filtrado de datos

```python
double_group["Age"].filter(lambda x: np.sum(x) > 2400)
```

## Transformación de variables

```python
zscore = lambda x: (x - x.mean())/x.std()
```

```python
z_group = double_group.transform(zscore)
import matplotlib.pyplot as plt

plt.hist(z_group["Age"])
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
