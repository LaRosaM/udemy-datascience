# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../datasets/movies/movies.csv", sep=";")
data
```

```python
movies = list(data.columns[1:])
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)
dd2 = distance_matrix(data[movies], data[movies], p=2)
dd10 = distance_matrix(data[movies], data[movies], p=10)
```

```python
# Distancia de Manhattan
d_dd1 = pd.DataFrame(dd1)
d_dd1
```

```python
# Distancia Euclidiana
d_dd2 = pd.DataFrame(dd2)
d_dd2
```

```python
# La distancia de Minkoski / Distance where p = 10
d_dd10 = pd.DataFrame(dd10)
d_dd10
```

# Clustering jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], 'average')
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
```

```python
Z = linkage(data[movies], 'ward')
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
```

```python
Z = linkage(data[movies], 'complete')
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
```

```python
Z = linkage(data[movies], 'single')
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
```
